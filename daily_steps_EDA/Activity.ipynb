```python
steps_table = Table().with_column('Steps',steps)
```

```python
averages = make_array() 
for i in np.arange(10000):
    resample = steps_table.sample()
    sample_average = np.mean(resample.column('Steps')) 
    averages = np.append(averages,sample_average)
```
