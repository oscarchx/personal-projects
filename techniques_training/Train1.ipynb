```python
##Random Walk
# Set the simulation parameters
mu = np.mean(modified_df['returns'])
vol = np.std(modified_df['returns'])
T = 252
S0 = 10

# Add one to the random returns
rand_rets = np.random.normal(mu,vol,T) + 1

# Forecasted random walk

forecasted_values = rand_rets.cumprod()*S0

# Plot the random walk
plt.plot(range(0, T), forecasted_values)
plt.show()
```

```python
# Loop through 100 simulations
for i in range(100):

    # Generate the random returns
    rand_rets = np.random.normal(mu, vol, T) + 1
    
    # Create the Monte carlo path
    forecasted_values = S0*(rand_rets).cumprod()
    
    # Plot the Monte Carlo path
    plt.plot(range(T), forecasted_values)

# Show the simulations
plt.show()
```

```python
# Aggregate the returns
sim_returns = []

# Loop through 100 simulations
for i in range(100):

    # Generate the Random Walk
    rand_rets = np.random.normal(mu, vol, T)
    
    # Save the results
    sim_returns.append(rand_rets)

# Calculate the VaR(99)
var_99 = np.percentile(sim_returns,1)
print("Parametric VaR(99): ", round(100*var_99, 2),"%")
```

```python
#95% CI

CI = [np.percentile(sim_returns,2.5),np.percentile(sim_returns,97.5)]
print(CI)
```

```python
###
start = date(2016,1,1)
end = date(2016,12,31)

# Set the ticker
ticker = ['AAPL','FB','GOOG','NFLX','K']

# Set the data source
data_source = 'iex'

# Import the stock prices
stock_prices5 = DataReader(ticker, data_source, start, end)

# Display and inspect the result
print(stock_prices5.head())
stock_prices5.info()
```

```python
stock_prices5.head()
```

```python
stock_prices5['close']
```

```python
close = stock_prices5['close']
columns = close.columns
print(columns)
```

```python
return_lists = []

for column in columns:
    returns = close[column].pct_change()
    return_lists.append(returns)
```

```python
return_lists
```

```python
returns = pd.DataFrame(return_lists).transpose()
returns
```

```python
# Finish defining the portfolio weights as a numpy array
portfolio_weights = np.array([0.20, 0.16, 0.14, 0.25, 0.25,0])

# Calculate the weighted stock returns
WeightedReturns = returns.mul(portfolio_weights, axis=1)

# Calculate the portfolio returns
returns['Portfolio'] = WeightedReturns.sum(axis=1)

# Plot the cumulative portfolio returns over time
CumulativeReturns = ((1+returns["Portfolio"]).cumprod()-1)
CumulativeReturns.plot()
plt.show()
```

```python
# How many stocks are in your portfolio?
numstocks = 5

# Create an array of equal weights across all assets
portfolio_weights_ew = np.repeat(1/numstocks,numstocks)

# Calculate the equally-weighted portfolio returns
returns['Portfolio_EW'] = returns.iloc[:,0:5].mul(portfolio_weights_ew, axis=1).sum(axis=1)
returns['Portfolio'].plot(subplots=False)
returns['Portfolio_EW'].plot(subplots=False)
```

```python
correlation_matrix = returns.corr()
print(correlation_matrix)
```

```python
# Calculate the covariance matrix
cov_mat = returns.cov()

# Annualize the co-variance matrix
cov_mat_annual = cov_mat*252

# Print the annualized co-variance matrix
print(cov_mat_annual)
```
